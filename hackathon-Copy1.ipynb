```python
for key, value in null_count.items():
    if value != 0:
        train[key] = train[key].astype(str)
        train[key].fillna(train[key].mode(), inplace=True)
```

```python
for key, value in null_count_s.items():
    if value != 0:
        test[key] = test[key].astype(str)
        test[key].fillna(test[key].mode(), inplace=True)
```

```python
from sklearn import preprocessing
def label_encoding(x_train):
    label_encoder = preprocessing.LabelEncoder()
    l = x_train.columns
    # Encode labels in column 'species'.
    for i in l:
        if isinstance(x_train[i][0], object):
            #print(i)
            x_train[i] = x_train[i].astype(str)
            x_train[i]= label_encoder.fit_transform(x_train[i])
```

```python
x_train = train.drop('Fault', axis=1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.feature_selection import RFE
```

```python

    model = LogisticRegression()
    rfe = RFE(estimator=model, n_features_to_select=1, step=1)
    rfe.fit(x_train, y_train)

    ranking_df = pd.DataFrame()
    ranking_df['Feature_name'] = x_train.columns
    ranking_df['Rank'] = rfe.ranking_

    ranked = ranking_df.sort_values(by=['Rank'])

    top_20 = list(ranked['Feature_name'])[:29]
    x_train_20 = x_train[top_20]
    
```

```python
x_train_20
```

```python
n_trees  = [628,633]
accu = []
for i in n_trees:
    clf = GradientBoostingClassifier(n_estimators=i, learning_rate=1.0,
     max_depth=1, random_state=0).fit(x_train_20, y_train)
    pred = clf.predict(x_train_20)
    acc = accuracy_score(pred, y_train)
    accu.append(acc)
    print("Number of trees =",i,"  Accuracy =",acc)
```
